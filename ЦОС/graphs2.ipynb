```python
import matplotlib.pyplot as plt
import numpy as np

signal_data = np.loadtxt('signal.txt')
restored_data = np.loadtxt('restored_signal.txt')

#  
time_signal = signal_data[:, 0]  #    
value_signal = signal_data[:, 1]  #   

time_restored = restored_data[:, 0]  #    
value_restored = restored_data[:, 1]  #   

#  
plt.figure(figsize=(10, 6))
plt.plot(time_signal, value_signal, label=' ', color='blue')
plt.plot(time_restored, value_restored, label=' ', color='red', linestyle='--')

plt.xlabel(' ()')
plt.ylabel('')
plt.title('    ')
plt.legend()
plt.grid(True)

plt.show()
```
