```python
import matplotlib.pyplot as plt
import numpy as np

signal_data = np.loadtxt('signal.txt')
restored_data = np.loadtxt('restored_signal.txt')

# Извлечение данных
time_signal = signal_data[:, 0]  # Время для исходного сигнала
value_signal = signal_data[:, 1]  # Значения исходного сигнала

time_restored = restored_data[:, 0]  # Время для восстановленного сигнала
value_restored = restored_data[:, 1]  # Значения восстановленного сигнала

# Построение графиков
plt.figure(figsize=(10, 6))
plt.plot(time_signal, value_signal, label='Исходный сигнал', color='blue')
plt.plot(time_restored, value_restored, label='Восстановленный сигнал', color='red', linestyle='--')

plt.xlabel('Время (с)')
plt.ylabel('Амплитуда')
plt.title('Сравнение исходного и восстановленного сигналов')
plt.legend()
plt.grid(True)

plt.show()
```
